```python
import numpy as np
import math

#for calculations DFT
#rotation along same direction
for i in range(0,360,15): #i in degree
    angle_degree = i
    a1_site1_degree = angle_degree
    a1_site2_degree = angle_degree + 180
    a1_site3_degree = angle_degree - 90
    a1_site4_degree = angle_degree - 90 + 180
    
    print("In the case of %.3f" % i)
    a1_site1_coor = 3*np.array([math.cos(math.radians(a1_site1_degree)), math.sin(math.radians(a1_site1_degree)), 0])
    a1_site2_coor = 3*np.array([math.cos(math.radians(a1_site2_degree)), math.sin(math.radians(a1_site2_degree)), 0])
    a1_site3_coor = 3*np.array([math.cos(math.radians(a1_site3_degree)), math.sin(math.radians(a1_site3_degree)), 0])
    a1_site4_coor = 3*np.array([math.cos(math.radians(a1_site4_degree)), math.sin(math.radians(a1_site4_degree)), 0])
    #magmom = np.array([a1_site1_coor, a1_site2_coor])
    magmom = (a1_site1_coor, a1_site2_coor, a1_site3_coor, a1_site4_coor)
    #print(magmom)
    #print("%s %s" %("24*0 12*0", magmom))
    print("24*0 12*0", *magmom, sep=' ',end=' 72*0')
    #print(*magmom, sep=' '),
    #print("72*0")
    #print(magmom)
    print("\n")
    
```
