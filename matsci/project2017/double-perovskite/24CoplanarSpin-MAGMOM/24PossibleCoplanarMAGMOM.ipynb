```python
import itertools as its
import numpy as np
    
FistSpinConf = np.array([[-2.121, -2.121, 0], [-2.121, 2.121, 0], [2.121, -2.121, 0], [2.121, 2.121, 0]])
SpinPermutation=list(its.permutations(FistSpinConf, 4))
# a
```

```python
f=open('24MAGMOM.txt', 'w')
for x in SpinPermutation:
    print('MAGMOM = 24*0 12*0', sep=' ', end=' ')
    print('MAGMOM = 24*0 12*0', sep=' ', end=' ', file=f)
    for y in x:
        for z in y:
            print(z, sep=' ', end=' ')
            print(z, sep=' ', end=' ', file=f)
    print('72*0')
    print('72*0', file=f)
        
f.close()
```
