```python
%matplotlib inline

# For the reference of the default color map of matplotlib, see
# https://matplotlib.org/users/dflt_style_changes.html

# This example is from CHAPTER 3 of the book "
# matplotlib for developpers",
# but I made many revisions

import matplotlib.style
import matplotlib as mpl
mpl.style.use('classic')
mpl.rcParams['figure.facecolor'] = '1'
#if choose the grey backgroud, use 0.75
mpl.rcParams['figure.figsize'] = [8,7]
mpl.rcParams['lines.linewidth'] = 1.5
mpl.rcParams['legend.fancybox'] = True
mpl.rcParams['legend.scatterpoints'] = 1 #scatterpoints,
#it's the numer of maker points in the legend when
#creating a legend entry for a scatter plot
mpl.rcParams["axes.formatter.useoffset"]=False #turn off the axis offset-values. 
# If on. the axis label will use a offset value by the side of axis

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#import pdf module
from matplotlib.backends.backend_pdf import PdfPages


a0=pd.read_csv('./0/dos/dossplit_f_ywfang/DOS0', header=None, delim_whitespace=True)
a20=pd.read_csv('./20/dos/dossplit_f_ywfang/DOS0', header=None, delim_whitespace=True)
a40=pd.read_csv('./40/dos/dossplit_f_ywfang/DOS0', header=None, delim_whitespace=True)
a60=pd.read_csv('./60/dos/dossplit_f_ywfang/DOS0', header=None, delim_whitespace=True)
#a0[0].head(10)


fig1 = plt.figure()
ax1_1=fig1.add_subplot(1,1,1)
#ax1_1.scatter(a[0], a[1], s=50, color='b', label='0 GPa$')
ax1_1.plot(a0[0], a0[1]/2, 'r-', label = '0 GPa') 
ax1_1.plot(a20[0], a20[1]/2, 'b-', label = '2 GPa') 
ax1_1.plot(a40[0], a40[1]/2, 'k-', label = '4 GPa') 
ax1_1.plot(a60[0], a60[1]/2, 'g-', label = '6 GPa') 
ax1_1.legend(loc="upper left")
#plt.xticks(x, SpinConfLabel, rotation='270')
plt.setp(ax1_1.get_xticklabels(), visible=True)
ax1_1.set_xlabel('Energy (eV)')
ax1_1.set_ylabel('Total density of states (states/eV/f.u.)')
ax1_1.locator_params(nbins=20, axis='y')
ax1_1.grid()
ax1_1.axhline(y=0, color='r', linestyle='-')
ax1_1.set_xlim([-1.0,1.0]) #set x-axis limit
ax1_1.set_ylim([0.0,4.5]) #set x-axis limit


pp1 = PdfPages('./TDOS.pdf')
pp1.savefig(fig1)

pp1.close()
```
