## 1.1 The total densities of states of LiOsO$_3$ under pressures

```python
%matplotlib inline

# For the reference of the default color map of matplotlib, see
# https://matplotlib.org/users/dflt_style_changes.html

# This example is from CHAPTER 3 of the book "
# matplotlib for developpers",
# but I made many revisions

import matplotlib.style
import matplotlib as mpl
mpl.style.use('classic')
mpl.rcParams['figure.facecolor'] = '1'
#if choose the grey backgroud, use 0.75
mpl.rcParams['figure.figsize'] = [8,6]
mpl.rcParams['lines.linewidth'] = 1.5
mpl.rcParams['legend.fancybox'] = True
mpl.rcParams['legend.scatterpoints'] = 1 #scatterpoints,
#it's the numer of maker points in the legend when
#creating a legend entry for a scatter plot
mpl.rcParams["axes.formatter.useoffset"]=False #turn off the axis offset-values. 
# If on. the axis label will use a offset value by the side of axis
mpl.rcParams["axes.linewidth"]  = 1.5 #change the boarder width
#plt.rcParams["axes.edgecolor"] = "0.15" #change the boarder color

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#import pdf module
from matplotlib.backends.backend_pdf import PdfPages


a0=pd.read_csv('./0/dos/dossplit_f_ywfang/DOS0', header=None, delim_whitespace=True)
a20=pd.read_csv('./20/dos/dossplit_f_ywfang/DOS0', header=None, delim_whitespace=True)
a40=pd.read_csv('./40/dos/dossplit_f_ywfang/DOS0', header=None, delim_whitespace=True)
a60=pd.read_csv('./60/dos/dossplit_f_ywfang/DOS0', header=None, delim_whitespace=True)
#a0[0].head(10)


fig1 = plt.figure()
ax1_1=fig1.add_subplot(1,1,1)
#ax1_1.scatter(a[0], a[1], s=50, color='b', label='0 GPa$')
ax1_1.plot(a0[0], a0[1]/2, 'r-', label = '0 GPa') 
ax1_1.plot(a20[0], a20[1]/2, 'b-', label = '2 GPa') 
ax1_1.plot(a40[0], a40[1]/2, 'y-', label = '4 GPa') 
ax1_1.plot(a60[0], a60[1]/2, 'g-', label = '6 GPa') 
ax1_1.legend(loc="lower right")
#plt.xticks(x, SpinConfLabel, rotation='270')
plt.setp(ax1_1.get_xticklabels(), visible=True)
ax1_1.set_xlabel('Energy (eV)')
ax1_1.set_ylabel('Total density of states (states/eV/f.u.)')
ax1_1.locator_params(nbins=20, axis='y')
ax1_1.grid()
ax1_1.axhline(y=0, color='r', linestyle='-')
ax1_1.set_xlim([-1.0,1.0]) #set x-axis limit
ax1_1.set_ylim([0.0,4.5]) #set x-axis limit
plt.axvline(x=0.0, color='k', linestyle='--') #add a vertical line



pp1 = PdfPages('./TDOS.pdf')
pp1.savefig(fig1)

pp1.close()

fig1 = plt.figure()
ax1_1=fig1.add_subplot(1,1,1)
#ax1_1.scatter(a[0], a[1], s=50, color='b', label='0 GPa$')
ax1_1.plot(a0[0], a0[1]/2, 'r-', label = '0 GPa') 
ax1_1.plot(a20[0], a20[1]/2, 'b-', label = '2 GPa') 
ax1_1.plot(a40[0], a40[1]/2, 'y-', label = '4 GPa') 
ax1_1.plot(a60[0], a60[1]/2, 'g-', label = '6 GPa') 
ax1_1.legend(loc="upper right")
#plt.xticks(x, SpinConfLabel, rotation='270')
plt.setp(ax1_1.get_xticklabels(), visible=True)
ax1_1.set_xlabel('Energy (eV)')
ax1_1.set_ylabel('Total density of states (states/eV/f.u.)')
ax1_1.locator_params(nbins=10, axis='y')
ax1_1.grid()
ax1_1.axhline(y=0, color='r', linestyle='-')
ax1_1.set_xlim([-10.0,5.0]) #set x-axis limit
ax1_1.set_ylim([0.0,10]) #set x-axis limit
plt.axvline(x=0.0, color='k', linestyle='--') #add a vertical line
pp1 = PdfPages('./TDOS-larger-energy-window.pdf')
pp1.savefig(fig1)

pp1.close()
```

# 4. merge the bond angle and gamma figure into one plot

```python
%matplotlib inline

# For the reference of the default color map of matplotlib, see
# https://matplotlib.org/users/dflt_style_changes.html

# This example is from CHAPTER 3 of the book "
# matplotlib for developpers",
# but I made many revisions

import matplotlib.style
import matplotlib as mpl
mpl.style.use('classic')
mpl.rcParams['figure.facecolor'] = '1'
#if choose the grey backgroud, use 0.75
mpl.rcParams['figure.figsize'] = [8,6]
mpl.rcParams['lines.linewidth'] = 1.5
mpl.rcParams['legend.fancybox'] = True
mpl.rcParams['legend.scatterpoints'] = 1 #scatterpoints,
#it's the numer of maker points in the legend when
#creating a legend entry for a scatter plot
mpl.rcParams["axes.formatter.useoffset"]=False #turn off the axis offset-values. 
# If on. the axis label will use a offset value by the side of axis
mpl.rcParams["axes.linewidth"]  = 1.5 #change the boarder width
#plt.rcParams["axes.edgecolor"] = "0.15" #change the boarder color

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas import DataFrame

#import pdf module
from matplotlib.backends.backend_pdf import PdfPages

data=pd.read_csv('./combine-angle-gamma/angle-pressure.txt', delim_whitespace=True)

frame = DataFrame(data)

with PdfPages('./angle-gamma.pdf') as pdf:
    ax1=frame.plot(['P'],secondary_y=['$\Gamma$-frequency'],color={'r','b'},
                  kind='line', marker='o',markersize=10,mark_right=False)
    
    ax1.set_ylabel('O1-Os-O4 bond angle ($^\circ$) of ${R3c}$ LiOsO$_3$',fontsize=16)
    ax1.set_xlabel('P (GPa)', fontsize = 16)
    ax1.right_ax.set_ylabel('$\Gamma$-frequency of $R$$c$ LiOsO$_3$',fontsize=16)
    ax1.right_ax.set_ylim([6.5,4.5]) #set y-axis limit (from bottom-6.5 to top-4.5)
    ax1.right_ax.yaxis.set_major_formatter(plt.FuncFormatter('{:.1f}$i$'.format)) #set the display format
    plt.xticks(np.arange(0, 13, 2))
    ax1.grid()
#    plt.axvline(x=5, color='k', linestyle='--') #add a vertical line
    pdf.savefig()
    plt.close()
```
